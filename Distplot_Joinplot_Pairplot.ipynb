#what is univariate and bivariate analysis?

when we are dealing with a single feature or independent variable and based on the same we are concerned in finding the target then it is called univariate analysis. 

when we are dealing with multiple features that is more than 1 feature and based on the same we are finding the output then it is called a bivariate analysis.

to visualise such kind of analysis we can use the seaborn library as it plays a vital role in analyzing such kind of data.



```
df= sns.load_dataset("tips")
```

```
df.head()
```

#Correlation with heatmap

Correlation basically tells the correlation between each and every feature with one another. A correlation map uses colored cells in a monochromatic scale to show a 2D correlation matrix between two discrete dimensions or event types.
It plays a pivotal role in feature selection.

Correlation can only be found out if our dataset contains integer or floats and not categorical features(strings)

```
df.corr()
```

```
df.dtypes
```

```
sns.heatmap(df.corr(),annot=True)
```

```
#so here basically say that if total bill is increasing then tips is also increasing and same goes for others because the value is positively correlated
```

```
#so based upon the degree of correlation we can select the features we want and which we don't want
```

Joint Plot

Used to plot histograms alongwith different types of plots

```
sns.jointplot(x="tip", y="total_bill", data=df, kind="hex")
```

```
#by looking at the above graph we can say that majority of the tip is between 2 and 4 dollars and bill payment is b/w 10 to 20 dollars
```

```
sns.jointplot(x="tip", y="total_bill", data=df, kind="reg")
```

```
#here above we can see the analysis using regression line
```

Pair plot

A pair plot is also known as a scatter plot in which one variable in the same data row is matched with another variables value.

```
sns.pairplot(df)
```

```
sns.pairplot(df, hue="sex")
```

```
sns.pairplot(df, hue="smoker")
```

```
df["smoker"].value_counts()
```

Dist Plot

distplot helps us to check the distributions of the columns feature

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

```
sns.distplot(df["tip"])
```

```
sns.distplot(df["tip"], kde=False, bins=10)
```

```
#so if kde is present then it will show the probability density and if it is not present then it will basically show the count that is integer values
```

```
#so these plots are used for numerical data types
```
