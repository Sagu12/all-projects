```
import io
import json
```

```
url_api= "https://api.ocr.space/parse/image"

_, compressed_image= cv2.imencode(".png", img, [1,90])

file_bytes= io.BytesIO(compressed_image)

result= requests.post(url_api,
              files= {"/content/tabular_data_1_the_perfect_format.png": file_bytes},
              data= {"apikey": "5262bf145888957", "isTable":"true"})

result= result.content.decode()

result= json.loads(result)

text_detected= result.get("ParsedResults")[0].get("TextOverlay").get("Lines")

text_detected

```
