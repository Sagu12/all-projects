```
import plotly.express as px
```

```
df= px.data.tips()
```

```
df.head()
```

```
t=df.loc[df.tip >= 2]
```

```
t.head()
```

# Bar plot

```
px.bar(data_frame=df, x="day", y="tip", title="tips given in different days")
```

```
px.bar(data_frame=df, x="tip", y="day", title="tips given in different days", orientation="h", color="size")
```

# Bar chart race plot


```
px.bar(data_frame=df,x="day",y="tip",animation_frame="tip", color="total_bill")
```

# Histogram Plot along with racing graph

```
px.histogram(data_frame=df, x="tip", title="tip distribution", color="day",animation_frame="tip")
```

# Box plot

```
px.box(data_frame=df,y="tip",color="day")
```

```
df.describe()
```

# Line plot

```
px.line(data_frame=df, x="day", y="tip",color="smoker")
```

# Scatter plot 3d

```
px.scatter_3d(data_frame=df,x="day",y="tip",z="total_bill",color="smoker")
```

# scatter plot with histogram and box plot


```
px.scatter(data_frame=df, x="tip",y="total_bill",color="day",marginal_y="histogram",marginal_x="box",trendline="ols", hover_name="smoker", log_x=True)
```
