```
fmri= sns.load_dataset("fmri")

fmri.tail()
```

```
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line")
```

```
#the blurred line represents the confidence level of the data and says that the range of the bold line lies within this confidence level
```

```
#to close the confidence level we need to do the following
```

```
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", ci=False)
```

```
#plotting with std deviation

sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", ci="sd")
```

```
#plotting the real data

sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", estimator=None)
```

```
#tips of how to standardize variables manually

#ypically, to standardize variables, you calculate the mean and standard deviation for a variable. Then, for each observed value of the variable, you subtract the mean and divide by the standard deviation
```

```
#Using hue, style and size for plotting the graph that is using semantically similar values to plot the graph for a better vizualization
```

```
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", hue="event")
```

```
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", hue="region", style="event", ci=None)
```

```
#removing the dashes and using markers for proper distinguishing
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", hue="region", style="event", ci=None, markers=True, dashes=False)
```

```
#plotting multiple line plots based on different subjects in the dataset of fmri in a graph and using only stim as the measuring parameter
```

```
sns.relplot(x="timepoint", y="signal",  kind="line",units="subject", hue="region", style="event", ci=None, markers=True, dashes=False, estimator=None, data= fmri.query("event=='stim'"))
```

```
dots= sns.load_dataset("dots").query("align=='dots'")
```

```
sns.relplot(x="time", y="firing_rate", data=dots, kind="line", hue="coherence", style="choice")
```
