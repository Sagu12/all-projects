# ***What is z score***?

# In gausian normal distribution we basically form a bell curve where we say that the mean is centered and the standard deviation lie adjacent to it. The bell curve also provides the density of the data that is being surrounded by the mean. Now to convert this gaussian distribution to standard normal distribution i.e, where mean is zero and standard deviation is 1 we use the following formula---------->

# **z score**---> **x(i) - u / Ïƒ**

# So this z score not only converts the gaussian distribution to standard normal distribution but also helps in finding out the density of the data within certain standard deviation from the mean.


#for example say we have the data points 1,2,3,4,5 and we find the mean of these data points as 3

#Now if we want to compute the z score of the point say 3 using the empirical formula we will get the following

#**z score= 3 - 3 / 1 = 0**

#so we got our output as zero and the same will be plotted in a graph for successive points (4,5,2,1).

#So z score basically helps in finding out the probability percentages of the data distributiobn away from the mean of the distribution curve.

# ***What is Normalization***?

# Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.

# ***What is Standardization***?

# Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.
